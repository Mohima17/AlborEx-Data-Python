```python
import os
import csv
import matplotlib.pyplot as plt
```

```python
datadir = "/home/ctroupin/Data/Alborex/raw/Alborex/Alborex/05-2014/termosal.proc"
datafile = os.path.join(datadir, "25052014.termosal.proc")
```

```python
dates, instrumentdates, T, S, sigmat, cond, fluor, Tremote= [], [], [], [], [], [], [], []
```
