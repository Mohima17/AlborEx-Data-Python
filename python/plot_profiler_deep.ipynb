Plot the deep values of the profiles

```python
import alborexdata
import alborexpaths
import os
import matplotlib.pyplot as plt
import numpy as np
from importlib import reload
```

## Plots
### PROVBIO
Read the data

```python
profilerbio = alborexdata.Profiler()
profilerbio.get_from_netcdf(alborexpaths.profilerfiles[1])
```

Plot all the profiles together

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111)
for i in range(0, len(profilerbio.dates)):
    plt.plot(profilerbio.salinity[i, :], profilerbio.depth[i, :], color="k", linewidth=1)
plt.gca().invert_yaxis()
plt.xlabel("Salinity", fontsize=14)
plt.ylabel("Depth (m)", fontsize=14)
plt.title("PROVBIO salinity profiles", fontsize=18)
plt.savefig(os.path.join(alborexpaths.figdir, "provbio_profiles"), dpi=300, bbox_inches="tight")
plt.ylim(500., 1000.)
plt.xlim(38.45, 38.55)
plt.gca().invert_yaxis()
plt.savefig(os.path.join(alborexpaths.figdir, "provbio_profiles_zoom"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
