```python
import shapefile
import matplotlib.pyplot as plt
import matplotlib.cm
import numpy as np
import os
```

```python
shapefileGC = "/home/ctroupin/DataOceano/ShapeFiles/GranCanaria/GranCanaria.shp"
shapefileGC = "/home/ctroupin/DataOceano/ShapeFiles/GobCan/MUNICIPIOS.shp"
```

```python
sf = shapefile.Reader(shapefileGC)
```

```python
shapes = sf.shapes()
len(shapes)
```

# Plot

```python
plt.cm.
```

```python
cmap = matplotlib.cm.get_cmap('autumn')
nshapes = len(shapes)
for i, municipios in enumerate(shapes):
    color = cmap(i/nshapes)
    a = np.array(municipios.points).T
    plt.plot(a[0], a[1], linewidth=.2, color=color)
    plt.text(a[0].mean(), a[1].mean(), str(i), fontsize=14, color=color)
plt.show()
plt.close()
```
